```python
import pandas as pd 
import numpy as np
```

```python
import tensorflow 
from tensorflow import keras 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense 
```

```python
from sklearn.datasets import make_circles
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
x , y = make_circles(n_samples=100 , noise=0.08 , random_state=1)
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],hue=y)
```

```python
model = Sequential()

model.add(Dense(128,input_dim=2,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='relu'))
```

```python
tensorflow.keras.optimizers.SGD(learning_rate=0.01,momentum=0.0,nesterov=False,name='SGD')
```

```python
model.compile(loss='binary_crossentropy',optimizer='SGD',metrics=['accuracy'])
```

```python
history = model.fit(x , y , batch_size=100 , epochs=200 , validation_split=0.2)
```
