```python
import numpy as np 
import pandas as pd 
```

```python
import tensorflow 
from tensorflow import keras 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense , Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
from sklearn.datasets import make_circles
import seaborn as sns 
```

```python
import matplotlib.pyplot as plt
```

```python
x , y = make_circles(n_samples=100 , noise=0.1 , random_state=1)
```

```python
y
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],hue=y)
```

```python
model = Sequential()

model.add(Dense(128,input_dim=2,activation='relu')) 
model.add(Dense(1,activation='relu')) 
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x , y , batch_size=100 , epochs=2000 , validation_split=0.2)
```

```python
plt.plot(history.history['loss'],label='train')
plt.plot(history.history['val_loss'],label='test')
plt.legend()
plt.show()
```

```python
model1 = Sequential()

model1.add(Dense(128,input_dim=2,activation='relu'))
model1.add(Dense(1,activation='relu'))
```

```python
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=40,
    verbose=1,
    mode = "auto",
    baseline=None,
)
```

```python
history1 = model1.fit(x ,y, batch_size=100,epochs=2000,validation_split=0.2 , callbacks=callback)
```

```python
history = history1
```

```python
plt.plot(history.history['loss'],label='train')
plt.plot(history.history['val_loss'],label='train')
plt.legend()
plt.show()
```
