```python
import pandas as pd 
import numpy as np
```

```python
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
import tensorflow 
from tensorflow import keras 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense 
```

```python
x,y = make_moons(100,noise=0.24,random_state=2)
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
model = Sequential()

model.add(Dense(128,input_dim=2,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='relu'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x,y,epochs=1000,validation_split=0.2,verbose=0)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
model1 = Sequential()

model1.add(Dense(128,activation='relu',input_dim=2,kernel_regularizer = tensorflow.keras.regularizers.l1(0.001)))
model1.add(Dense(128,activation='relu',kernel_regularizer = tensorflow.keras.regularizers.l1(0.001)))
model1.add(Dense(1,activation='relu'))
```

```python
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history1 = model1.fit(x,y,epochs=1000,validation_split=0.2,verbose=0)
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```
