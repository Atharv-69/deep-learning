```python
import pandas as pd 
import numpy as np
```

```python
import tensorflow 
from tensorflow import keras 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv("D:\placement (2).csv")
```

```python
df.head()
```

```python
x=df.drop(['placed'],axis=1)
y = df['placed']
```

```python
model = Sequential()

model.add(Dense(128,input_dim=2))
model.add(Dense(128))
model.add(Dense(128))
model.add(Dense(1))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x,y,epochs=500,validation_split=0.2)
```

```python
model1 = Sequential()

model1.add(Dense(128,activation='relu',input_dim=2))
model1.add(Dense(128,activation='relu'))
model1.add(Dense(128,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history1 = model1.fit(x,y,epochs=500,validation_split=0.2)
```
